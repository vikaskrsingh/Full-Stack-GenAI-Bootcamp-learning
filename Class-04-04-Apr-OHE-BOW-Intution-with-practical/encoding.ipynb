```python
from sklearn.preprocessing import OneHotEncoder
import numpy
```

```python
document = ["my name is sunny and I love AI"]
```

```python
documents = ["people watch movie", 
             "people watch cricket", 
             "people like movie", 
             "people like cricket"]
```

```python
tokens = [sentence.lower().split() for sentence in document]
```

```python
tokens
```

```python
all_words = [[word] for sentence in tokens for word in sentence]
```

```python
all_words
```

##### sparse_output=False means the encoder returns a NumPy array instead of a sparse matrix. This makes the output easier

https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.OneHotEncoder.html

```python
encoder = OneHotEncoder(sparse_output=False)
```

```python
encoder.fit(all_words)
```

```python
encoder.categories_
```

```python
print("Vocabulary:", encoder.categories_[0])
```

```python
tokens
```

```python
for sentence in tokens:
    print(sentence)
```

```python
for sentence in tokens:
    encoded_sentence = encoder.transform([[word] for word in sentence])
    print("\nSentence:", sentence)
    print(encoded_sentence)
```

# BOW

###### https://scikit-learn.org/0.15/modules/generated/sklearn.feature_extraction.text.CountVectorizer.html

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
documents = ["people watch movie and watch movie again", 
             "people watch cricket and watch cricket",
             "people like movie and like movie a lot",
             "people like cricket"]
```

```python
document=["divesh is a good mentor he teaches us very well he teaches the project"]
```

AIM: Covert: Data: Number

```python
bow = CountVectorizer()
```

```python
bow_matrix =bow .fit_transform(document)
```

```python
bow_matrix
```

```python
print("BOW Matrix:\n", bow_matrix.toarray())
```

```python
bow.get_feature_names_out()
```

```python
print("Vocabulary:", bow.get_feature_names_out())
```

```python
bow.fit(documents)
```

```python
documents
```

```python
bow.transform(documents).toarray()
```

```python
document = ["lion is the king of jungle"]
```

```python
bow.transform(document)
```

# TF-IDF Practical

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
documents = ["people watch movie and watch movie again", 
             "people watch cricket and watch cricket",
             "people like movie and like movie a lot",
             "people like cricket"]
```

```python
tf_idf = TfidfVectorizer()
```

```python
tf_idf_vector = tf_idf.fit_transform(documents)
```

```python
tf_idf_vector
```

```python
tf_idf_vector.toarray()
```

```python
for vector in tf_idf_vector.toarray():
    print(vector)
```

```python
tf_idf.get_feature_names_out()
```

```python
print("Vocabulary:", tf_idf.get_feature_names_out())
```
